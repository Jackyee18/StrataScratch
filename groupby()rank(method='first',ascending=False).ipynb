# Top 3 Facilities

Find the top 3 facilities for each owner. The top 3 facilities can be identified using the highest average score for each owner name and facility address grouping.
The output should include 4 columns: owner name, top 1 facility address, top 2 facility address, and top 3 facility address. Order facilities with the same score alphabetically.

```python
import pandas as pd
import numpy as np
q=pd.read_csv('los_angeles_restaurant_health_inspections.csv')
pd.set_option('display.max_rows',500)
```

```python
# find the average score of each facility according to each owner_name and facility address
q.groupby(['owner_name','facility_address'])['score'].mean().reset_index()
q=q.groupby(['owner_name','facility_address'])['score'].mean().reset_index()
```

```python
# add the rank of score of every facility address of each owner name
q['score_rank']=q.groupby('owner_name')['score'].rank(method='first',ascending=False)
```

```python
# create the top 1 score rank, top 2 score rank and top 3 score rank
q1=q[q['score_rank']==1].drop(columns=['score','score_rank'])
q2=q[q['score_rank']==2].drop(columns=['score','score_rank'])
q3=q[q['score_rank']==3].drop(columns=['score','score_rank'])
```

```python
# merge 3 table to get top 3 facility of each owner name 
q=q1.merge(q2,how='left',on='owner_name').merge(q3,how='left',on='owner_name')
```

```python
# rename the columns to generate the result
q.rename(columns={'facility_address_x':'top_1_facility_address','facility_address_y':'top_2_facility_address','facility_address':'top_3_facility_address'})
```
