# Find The Most Profitable Location


Find the most profitable location. Write a query that calculates the average signup duration and average transaction amount for each location, and then compare these two measures together by taking the ratio of the average transaction amount and average duration for each location.


Your output should include the location, average duration, average transaction amount, and ratio. Sort your results from highest ratio to lowest.

```python
import pandas as pd
q=pd.read_table('signups.txt')
q[['signup_start_date','signup_stop_date']]=q[['signup_start_date','signup_stop_date']].apply(pd.to_datetime)
```

```python
q['signup_duration_days']=(q.signup_stop_date-q.signup_start_date).dt.days
```

```python
t=pd.read_table('transactions.txt')
t.transaction_start_date=t.transaction_start_date.apply(pd.to_datetime)
t.groupby('signup_id').amt.mean().reset_index(name='total_amt')
```

```python
final=q[['location','ignup_id','signup_duration_days']].merge(t.groupby('signup_id').amt.sum().reset_index(name='total_amt'),left_on='ignup_id',right_on='signup_id')
final=final[['location','total_amt','signup_duration_days']]
final
```

```python
final['amt_days_ratio']=final.total_amt/final.signup_duration_days
```

```python
final.sort_values('amt_days_ratio',ascending=False)
```
