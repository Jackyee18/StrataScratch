# Product Market Share



Write a query to find the Market Share at the Product Brand level for each Territory, for Time Period Q4-2021. Market Share is the number of Products of a certain Product Brand brand sold in a territory, divided by the total number of Products sold in this Territory.
Output the ID of the Territory, name of the Product Brand and the corresponding Market Share in percentages. Only include these Product Brands that had at least one sale in a given territory.

```python
import pandas as pd
sales=pd.read_csv('fct_customer_sales.csv')
cus_ter=pd.read_csv('map_customer_territory.csv')
dim=pd.read_csv('dim_product.csv')
pd.set_option('display.max_row',500)
sales
```

```python
sales['order_date']=pd.to_datetime(sales.order_date)
```

```python
q4_2021_sales=sales[(sales.order_date.dt.quarter==4)&(sales.order_date.dt.year==2021)]
q4_2021_sales
```

```python
q=pd.merge(q4_2021_sales,cus_ter,left_on='ust_id',right_on='cust_id').merge(dim)
q
```

Write a query to find the Market Share at the Product Brand level for each Territory, for Time Period Q4-2021. Market Share is the number of Products of a certain Product Brand brand sold in a territory, divided by the total number of Products sold in this Territory. Output the ID of the Territory, name of the Product Brand and the corresponding Market Share in percentages. Only include these Product Brands that had at least one sale in a given territory.

```python
q1=q.groupby(['prod_brand','territory_id']).order_value.sum().reset_index(name='total_orders')
q1
```

```python
q2=q.groupby('prod_brand').order_value.sum().reset_index(name='total_orders')
q2
```

```python
q3=q1.merge(q2,on='prod_brand').rename(columns={'total_orders_x':'total_order_per_ter','total_orders_y':'total_order_per_brand'})
q3
```

```python
q3['percentage']=(q3.total_order_per_ter/q3.total_order_per_brand*100).round()
q3
```
