# Common Letters



Find the top 3 most common letters across all the words from both the tables (ignore filename column). Output the letter along with the number of occurrences and order records in descending order based on the number of occurrences.

```python
import pandas as pd
import numpy as np
q1=pd.read_csv('google_file_store.csv')
q1
```

```python
q2=pd.read_csv('google_word_lists.csv')
q2
```

```python
# concat the columns which contains the contents
q=pd.concat([q1.contents,q2.words1,q2.words2])
q
```

```python
# set a new dataframe which contain all content
q=pd.DataFrame(q).rename(columns={0:'content'})
q
```

```python
# combine multiple rows into a cell
content=q.content.str.cat(sep=';')
content
```

```python
# create a list of letter (from 'a' to 'z')
import string
letter_list=list(string.ascii_lowercase)
letter_list=pd.DataFrame(letter_list,columns=['letter'])
letter_list
```

```python
# create a list to contain results
number=[]

for i in letter_list.letter.values: # get iterates letters
    x=content.count(i) # find the total occurrences of content of each letter
    number.append(x) # save the result into a list
number
```

```python
# add new column with the above result: the total count number
letter_list['total_count']=number
```

```python
# final results:
letter_list.nlargest(3,'total_count')
```
