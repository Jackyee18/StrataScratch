# Delivering and Placing Orders


Check if there is a correlation between average total order value and average time in minutes between placing the order and delivering the order per restaurant.



```python
import pandas as pd
q=pd.read_table('doordash_delivery.txt')
q[['customer_placed_order_datetime','placed_order_with_restaurant_datetime','driver_at_restaurant_datetime','delivered_to_consumer_datetime']]=q[['customer_placed_order_datetime','placed_order_with_restaurant_datetime','driver_at_restaurant_datetime','delivered_to_consumer_datetime']].apply(pd.to_datetime)
q
```

```python
q['processing_time_minute']=(q.delivered_to_consumer_datetime-q.placed_order_with_restaurant_datetime).dt.seconds/60
```

```python
q.groupby('restaurant_id').processing_time_minute.mean().reset_index(name='avg_process_time')
```

```python
q.groupby('restaurant_id').order_total.sum().reset_index(name='avg_sales')
```

```python
final=q.groupby('restaurant_id').processing_time_minute.mean().reset_index(name='avg_process_time').merge(q.groupby('restaurant_id').order_total.sum().reset_index(name='avg_sales'))
```

```python
final.avg_process_time.corr(final.avg_sales)
```
