# Most Expensive And Cheapest Wine


Find the cheapest and the most expensive variety in each region. Output the region along with the corresponding most expensive and the cheapest variety. Be aware that there are 2 region columns, the price from that row applies to both of them.


Note: The results set contains no ties.

```python
import pandas as pd
import numpy as np
q=pd.read_csv('winemag_p1.csv',encoding='latin-1')
q
```

```python
# add the new column which contains max price of each variety & country 
q['most_exp_variety_price']=q.groupby(['country','variety']).price.transform(max)
```

```python
# add the new column which contains min price of each variety & country 
q['cheapest_variety_price']=q.groupby(['country','variety']).price.transform(min)
```

```python
# filter out the most expensive variety in dataset 
q1=q.loc[(q.price==q['most_exp_variety_price']),['country','variety','price']]
```

```python
# filter out the cheapest variety in dataset
q2=q.loc[(q.price==q['cheapest_variety_price']),['country','variety','price']]
```

```python
pd.merge(q1,q2,on=['country','variety']).rename(columns={'price_x':'most_variety_price','price_y':'cheapest_variety_price'})
```

```python
# other solution to check again:
# group by multiple columns with multiple aggregation
q.groupby(['country','variety']).agg({'price':['min','max']})
```
