Find all number pairs whose first number is smaller than the second one and the product of two numbers is larger than 11.
Output both numbers in the combination.

```python
import pandas as pd
import numpy as np
q=pd.read_csv('transportation_numbers.csv')
pd.set_option('display.max_row',500)
q
```

```python
# find the second_number which is larger than first_number
second_number=[]
for i in q.number.values:
    for j in q.number.values:
        if i<j:
            second_number.append(j)
second_number
```

```python
# find the first_number which is according to the aboved second_number
first_number=[]
for i in q.number.values:
    for j in q.number.values:
        if i<j:
            first_number.append(i)
first_number
```

```python
# create a dataframe with 2 variables
q=pd.DataFrame({'first_number':first_number,'second_number':second_number})
q
```

# Find all number pairs whose first number is smaller than the second one and the product of two numbers is larger than 11

Find all number pairs whose first number is smaller than the second one and the product of two numbers is larger than 11.
Output both numbers in the combination.

```python
import pandas as pd
q=pd.read_csv('transportation_numbers.csv')
# find the second_number which is larger than first_number
second_number=[]
for i in q.number.values:
    for j in q.number.values:
        if i<j:
            second_number.append(j)
# find the first_number which is according to the aboved second_number
first_number=[]
for i in q.number.values:
    for j in q.number.values:
        if i<j:
            first_number.append(i)
# create a dataframe with 2 variables
q=pd.DataFrame({'first_number':first_number,'second_number':second_number})
# filter the pairs of 2 number which the product of them is greater than 11
q[(q.first_number*q.second_number)>11]
```
