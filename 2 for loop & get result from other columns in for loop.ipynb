# Find fare differences on the Titanic using a self join


Find the average absolute fare difference between a specific passenger and all passengers that belong to the same pclass,  both are non-survivors and age difference between two of them is 5 or less years. Do that for each passenger (that satisfy above mentioned coniditions). Output the result along with the passenger name.

```python
import pandas as pd
import numpy as np
q=pd.read_csv('titanic.csv')
# filter out only the survivors in dataset and the ones which contain age values & only extract 3 necessary columns
q=q.loc[(q.survived==0)&(q.age.notnull()),['name','age','fare','pclass']] 
q
```

```python
# reset index for new dataframe
q=q.reset_index().drop(columns='index')
q
```

```python
# create an empty list to contain results 
first_name=[]
second_name=[]
first_age=[]
second_age=[]
first_fare=[]
second_fare=[]
first_class=[]
second_class=[]

# create 2 loop to compare each other and get result from 2 tables

for i in range(len(q)): # create a range of number according to the length of the tables
    for j in range(len(q)):
        if abs(q.iloc[i,1] - q.iloc[j,1]) <=5: # codition: age difference between two of them is 5 or less years 
            first_name.append(q.iloc[i,0]) # add name of the first passenger
            second_name.append(q.iloc[j,0]) # add name of the second passenger
            first_age.append(q.iloc[i,1]) # add age of the first passenger
            second_age.append(q.iloc[j,1]) # add age of the second passenger
            first_fare.append(q.iloc[i,2]) # add fare of the first passenger
            second_fare.append(q.iloc[j,2]) # add fare of the second passenger
            first_class.append(q.iloc[i,3]) # add class of the first passenger
            second_class.append(q.iloc[j,3]) # add class of the second passenger
# create a dataframe of total pairs
pairs=pd.DataFrame({'first_name':first_name,'first_age':first_age,'first_fare':first_fare,'first_class':first_class,
                    'second_name':second_name,'second_age':second_age,'second_fare':second_fare,'second_class':second_class})
pairs
```

```python
# find 2 different person and only  a specific passenger and all passengers
final=pairs[(pairs.first_name!=pairs.second_name)&(pairs.first_name==pairs.first_name[0])]
final
```

```python
final['fare_diff']=abs(final.second_fare-final.first_fare)
```

```python
final
```

```python
final.groupby('second_class').fare_diff.mean()
```
