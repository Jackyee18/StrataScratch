# Find the average number of friends a user has


Find the average number of friends a user has.



```python
import pandas as pd
import numpy as np
q=pd.read_csv('google_friends_network.csv')
q
```

```python
# find the total friend of each users
q.groupby('user_id')['friend_id'].count().reset_index(name='total_friends')
```

```python
# fidn the average number of friends a user has
q.groupby('user_id')['friend_id'].count().reset_index(name='total_friends').total_friends.mean()
```
