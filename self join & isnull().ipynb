# Norwegian Alpine Skiers


Find all Norwegian alpine skiers who participated in 1992 but didn't participate in 1994. Output unique athlete names.



```python
import pandas as pd
import numpy as np
q=pd.read_csv('olympics_athletes_events.csv')
pd.set_option('display.max_row',500)
q
```

```python
# check the values of 'sport' column
q.sport.unique()
```

```python
# find the total events of each athletes
q[(q.sport=='Alpine Skiing')].groupby('year')['name'].value_counts().reset_index(name='total')
```

```python
# find the total 'Alpine Skiing' events of each athletes in 1994
q94=q[(q.sport=='Alpine Skiing')&(q.year==1994)][['name']].value_counts().reset_index(name='total')
```

```python
# find the total 'Alpine Skiing' events of each athletes in 1992
q92=q[(q.sport=='Alpine Skiing')&(q.year==1992)][['name']].value_counts().reset_index(name='total')
```

```python
# merge 2 table
in_92_and_94=pd.merge(q92,q94,how='left',on='name')
in_92_and_94
```

```python
# find which athletes participating in 92 but in 94: total events in 94(total_y) equal null
in_92_and_94.loc[in_92_and_94['total_y'].isnull(),'name']
```
