# Find the first and last inspection dates for vermin infestations per municipality

Find the first and last inspections for vermin infestations per municipality.
Output the result along with the business postal code.

```python
import pandas as pd
import numpy as np
q=pd.read_csv('sf_restaurant_health_violations.csv',encoding='latin-1')
q
```

```python
# change the data type into datetime
q.inspection_date=q.inspection_date.apply(pd.to_datetime)
```

```python
# find the violation which contain 'vermin'
q=q[q.violation_description.str.contains('vermin',na=False)].iloc[:,[5,-6]]
q
```

```python
# rank the inspection date of each business postal code in ascending order
q['first_inspection_date']=q.groupby('business_postal_code').inspection_date.rank()
```

```python
# rank the inspection date of each business postal code in descending order
q['last_inspection_date']=q.groupby('business_postal_code').inspection_date.rank(ascending=False)
```

```python
# find the first inspection date of each business postal codes
first_inspection_date=q[(q.first_inspection_date==1)].drop(columns=['first_inspection_date','last_inspection_date'])
```

```python
# find the last inspection date of each business postal codes
last_inspection_date=q[(q.last_inspection_date==1)].drop(columns=['first_inspection_date','last_inspection_date'])
```

```python
# merge 2 aboved tables together:
pd.merge(first_inspection_date,last_inspection_date,on='business_postal_code').rename(columns={
    'inspection_date_x':'first_inspection_date','inspection_date_y':'last_inspection_date'})
```
