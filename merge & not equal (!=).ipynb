# Find countries that are in winemag_p1 dataset but not in winemag_p2

Find countries that are in winemag_p1 dataset but not in winemag_p2.
Output distinct country names.
Order records by the country in ascending order.

```python
import pandas as pd
import numpy as np
q1=pd.read_csv('winemag_p1.csv',encoding='latin1')
q1=q1[['id','country']]
q1
```

```python
q2=pd.read_csv('winemag_p2.csv',encoding='latin1')
q2=q2[['id','country']]
q2
```

```python
q3=pd.merge(q1,q2,on='id',how='left')
q3[q3['country_x']!=q3['country_y']].drop(columns='country_y')
q4=q3[q3['country_x']!=q3['country_y']].drop(columns='country_y')
```

```python
q4.country_x.unique()
```
