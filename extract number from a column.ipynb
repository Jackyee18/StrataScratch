# Points Rating Of Wines Over Time


Find the average points difference between each and previous years starting from the year 2000. Output the year, average points, previous average points, and the difference between them.
If you're unable to calculate the average points rating for a specific year, use an 87 average points rating for that year (which is the average of all wines starting from 2000).

```python
import pandas as pd
import numpy as np
q=pd.read_csv('winemag_p2.csv')
```

```python
## extract year from the title
# other pattern: q.title.str.extract('([(0-9)]+)')
q['year']=q.title.str.extract('(\d{4})')
```

```python
# check the result of the year column
q.year.unique()
```

```python
# replace the NaN values with 2000
q.loc[q.year.isnull(),'year']=2000
```

```python
# check again if the value has changed
q.year.unique()
```

```python
# replace the points of 2000 with 87
q.loc[q.year.isnull(),'points']=87
```

```python
# find the avarage point of each years
q=q.groupby('year')['points'].mean().reset_index()
# change the type of year column before sorting values
q.year=q.year.astype(int)
q=q.sort_values('year',ascending=False)
q
```

```python
# add the preivous points and point diff
q['previous_points']=q.points.shift().round()
q['points_diff']=q.points.diff().round()
q
```
