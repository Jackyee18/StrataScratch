# Videos Removed on Latest Date


For each unique user in the dataset, find the latest date when their flags got reviewed. Then, find total number of distinct videos that were removed on that date (by any user).
Output the the first and last name of the user (in two columns), the date and the number of removed videos. Only include these users who had at least one of their flags reviewed by Youtube. If no videos got removed on a certain date, output 0.

```python
import pandas as pd
uf=pd.read_csv('user_flags.csv')
fr=pd.read_csv('flag_review.csv')
```

```python
q=uf.merge(fr)
q
```

```python
q.reviewed_date=pd.to_datetime(q.reviewed_date)
```

```python
video_review=q[q.reviewed_by_yt==True]
video_review
```

```python
latest_review_per_video=video_review.groupby(['user_firstname','user_lastname']).reviewed_date.max().reset_index(name='latest_review_date')
latest_review_per_video
```

```python
removed_each_date=q[q.reviewed_outcome=='REMOVED'].groupby('reviewed_date').reviewed_outcome.count().reset_index(name='removed_count')
removed_each_date
```

```python
latest_review_per_video.merge(removed_each_date,left_on='latest_review_date',right_on='reviewed_date',how='left').fillna(0).drop(columns='reviewed_date')
```
