# Popular Posts


The column 'perc_viewed' in the table 'post_views' denotes the percentage of the session duration time the user spent viewing a post. Using it, calculate the total time that each post was viewed by users. Output post ID and the total viewing time in seconds, but only for posts with a total viewing time of over 5 seconds.

```python
import pandas as pd
s=pd.read_table('user_sessions.txt')
s[['session_starttime','session_endtime']]=s[['session_starttime','session_endtime']].apply(pd.to_datetime)
s
```

```python
s['session_duration']=(s.session_endtime-s.session_starttime).dt.seconds
```

```python
p=pd.read_csv('post_views.csv')
```

```python
q=pd.merge(p,s)
q=q.iloc[:,[0,1,2,-1]]
q
```

```python
q['post_duration']=q.perc_viewed*q.session_duration/100
```

```python
post_view=q.groupby('post_id').post_duration.sum().reset_index(name='total_view_in_second')
post_view
```

```python
post_view[post_view.total_view_in_second>5]
```
