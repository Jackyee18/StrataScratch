# Negative Reviews in New Locations



Find stores that were opened in the second half of 2021 with more than 20% of their reviews being negative. A review is considered negative when the score given by a customer is below 5. Output the names of the stores together with the ratio of negative reviews to positive ones.

```python
import pandas as pd
r=pd.read_csv('instacart_reviews.csv')
r
```

```python
s=pd.read_csv('instacart_stores.csv')
s.opening_date=pd.to_datetime(s.opening_date)
```

```python
s[s.opening_date.dt.month>6]
```

```python
r[r.score<5].groupby('store_id').id.count().reset_index(name='total_negative_reviews').merge(s[s.opening_date.dt.month>6],left_on='store_id',right_on='id')
```

## OTHER SOLUTION

```python
review_type=[]
for i in r.score.values:
    if i <5:
        review_type.append('negative')
    else:
        review_type.append('positive')
r['review_type']=review_type
```

```python
r[r.review_type=='negative'].groupby('store_id').id.count().reset_index(name='total_negative_reviews').merge(s[s.opening_date.dt.month>6],left_on='store_id',right_on='id')
```
