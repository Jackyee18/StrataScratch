# Find the oldest survivor per passenger class



Find the oldest survivor of each passenger class.
Output the name and the age of the survivor along with the corresponding passenger class.
Order records by passenger class in ascending order.

```python
import pandas as pd
import numpy as np
q=pd.read_csv('titanic.csv')
q
```

```python
# filter out all the survivors in dataset
q=q[q.survived==1]
```

```python
# add the column of max age each classes
q['max_age_each_class']=q.groupby('pclass').age.transform('max')
q
```

```python
# Find the oldest survivor of each passenger class.
q.loc[q.age==q['max_age_each_class'],['name','pclass','age']]
```
