```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows',500)
q=pd.read_csv('online_retail.csv',encoding='latin-1')
q.invoicedate=pd.to_datetime(q.invoicedate)
q
```

# Best Selling Item


Find the best selling item for each month (no need to separate months by year) where the biggest total invoice was paid. The best selling item is calculated using the formula (unitprice * quantity). Output the description of the item along with the amount paid.



```python
q['total_amount']=q.quantity*q.unitprice
```

```python
q1=q.groupby([q.invoicedate.dt.month,'stockcode']).total_amount.sum().reset_index()
```

```python
q1
```

```python
q1['rank_total_amount']=q1.groupby('invoicedate')['total_amount'].rank(ascending=False)
```

```python
q1=q1[q1.rank_total_amount==1]
```

```python
q2=q[['stockcode','description']].value_counts().reset_index()
```

```python
pd.merge(q1,q2,on='stockcode').drop(columns=['rank_total_amount',0])
```
