# Average Time Between Steps


Find the average time (in seconds), per product, that needed to progress between steps. You can ignore products that were never used. Output the feature id and the average time.

```python
import pandas as pd
import numpy as np
q=pd.read_csv('facebook_product_features_realizations.csv')
q
```

```python
# change datatype of timestamp into datetime
q.timestamp=pd.to_datetime(q.timestamp)
```

```python
# find the timestamp of next step of each user_id of each feature
q['last_timestamp']=q.groupby(['feature_id','user_id']).timestamp.shift()
q
```

```python
# find the time diff
q.timestamp-q.last_timestamp
```

```python
# change into seconds
q['time_between_step']=(q.timestamp-q.last_timestamp).dt.seconds
```

```python
# find the average time between steps of each user of each feature
q.groupby(['feature_id','user_id']).time_between_step.mean().reset_index(name='avg_time')
```

```python
# find the average time of each feature
q.groupby('feature_id',).time_between_step.mean().reset_index(name='avg_time')
```
