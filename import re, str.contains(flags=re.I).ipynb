# 3rd Most Reported Health Issues


Each record in the table is a reported health issue and its classification is categorized by the facility type, size, risk score which is found in the pe_description column.


If we limit the table to only include businesses with Cafe, Tea, or Juice in the name, find the 3rd most common category (pe_description). Output the name of the facilities that contain 3rd most common category.

```python
import pandas as pd
import numpy as np
q=pd.read_csv('los_angeles_restaurant_health_inspections.csv')
q
```

```python
import re # import the regex library specialized in string
q=q[q.facility_name.str.contains('Cafe|Tea|Juice',flags=re.I)]
q
```

```python
# find the total issues per categories (pe_description)
q1=q.groupby('pe_description').erial_number.count().reset_index(name='total_counts')
q1
```

```python
# find the 3rd most health issues
q1.nlargest(3,'total_counts').iloc[2,:][0]
```

```python
# find the name of facilities that contain 3rd most common category
q[q.pe_description==q1.nlargest(3,'total_counts').iloc[2,:][0]]
```
