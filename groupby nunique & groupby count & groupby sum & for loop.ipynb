# WFM Brand Segmentation based on Customer Activity


WFM would like to segment the customers in each of their store brands into Low, Medium, and High segmentation. The segments are to be based on a customer's average basket size which is defined as (total sales / count of transactions), per customer.


The segment thresholds are as follows:



If average basket size is more than 30$, then Segment is “High”.

If average basket size is between 20 and 30$, then Segment is “Medium”.

If average basket size is less than 20$, then Segment is “Low”.


Summarize the number of unique customers, the total number of transactions, total sales, and average basket size, grouped by store brand and segment for 2017.


Your output should include the brand, segment, number of customers, total transactions, total sales, and average basket size.

```python
import pandas as pd
q=pd.read_table('wfm_transactions.txt')
pd.to_datetime(q.transaction_date)
q
```

```python
customer_per_store=q.groupby('store_id').customer_id.nunique().reset_index(name='total_customers')
```

```python
sales_per_store=q.groupby('store_id').sales.sum().reset_index(name='total_sales')
```

```python
transactions_per_store=q.groupby('store_id').transaction_id.count().reset_index(name='total_transactions')
```

```python
final=pd.merge(customer_per_store,transactions_per_store).merge(sales_per_store)
```

```python
final['basket_size']=(final.total_sales/final.total_customers/final.total_transactions).round(1)
```

```python
basket_size=[]
for i in final.basket_size.values:
    if i>30:
        basket_size.append('High')
    elif i>20:
        basket_size.append('Medium')
    else:
        basket_size.append('Low')
basket_size
```

```python
final['segment']=basket_size
```

```python
final.merge(pd.read_table('wfm_stores.txt'))
```
