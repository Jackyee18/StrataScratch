# Number Of Acquisitions

Find the number of acquisitions that occurred in each quarter of each year. Output the acquired quarter in YYYY-Qq format along with the number of acquisitions and order results by the quarters with the highest number of acquisitions first.


```python
q.acquired_quarter=pd.to_datetime(q.acquired_quarter)
```

```python
q.columns
```

```python
q1=q.groupby('acquired_quarter').acquired_at.count().reset_index().rename(columns={'acquired_at':'total_acquisitions'})
q1
```

```python
# Change type and substring the date column:q1.acquired_quarter.astype(str).str[:4]
# Change type and substring and replace values on multiple conditions
q1.acquired_quarter.astype(str).str[:4]+'-'+q1.acquired_quarter.astype(str).str[5:7].replace({'01':'Q1','04':'Q2','07':'Q3','10':'Q4'})
```
