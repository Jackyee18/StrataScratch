# Views Per Keyword


Create a report showing how many views each keyword has. Output the keyword and the total views, and order records with highest view count first.



```python
import pandas as pd
import numpy as np
q=pd.read_csv('facebook_posts.csv') # read the fb table
q
```

```python
# read fb post view file
v=pd.read_csv('facebook_post_views.csv')
v
```

```python
# find the total views of each posts
total_v=v.groupby('post_id').viewer_id.count().reset_index(name='total_views')
```

```python
# find the total view of each keyword by merging with fb post table
pd.merge(q,total_v,on='post_id').iloc[:,[3,5]]
```
