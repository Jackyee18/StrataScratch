# More Than 100 Dollars



For each month of 2021, calculate what percentage of restaurants, out of these that fulfilled any orders in a given month, fulfilled more than 100$ in monthly sales?

```python
import pandas as pd
do=pd.read_csv('delivery_orders.csv')
pd.set_option('display.max_row',500)
do
```

```python
do['order_placed_time']=pd.to_datetime(do.order_placed_time)
```

```python
do['month']=do.order_placed_time.dt.month
```

```python
do2021=do[do.order_placed_time.dt.year==2021]
```

```python
ov=pd.read_csv('order_value.csv')
ov
```

```python
q=pd.merge(do2021,ov,how='left')
q
```

```python
total_sales_each_month_per_restaurant=q.groupby(['restaurant_id','month']).sales_amount.sum().reset_index(name='total_sales')
total_sales_each_month_per_restaurant
```

```python
flag=[]
for i in total_sales_each_month_per_restaurant.total_sales.values:
    if i > 100:
      flag.append('Y')
    else:
        flag.append('N')
flag
```

```python
total_sales_each_month_per_restaurant['flag']=flag
```

```python
total_sales_each_month_per_restaurant
```

```python
total_sales_each_month_per_restaurant[total_sales_each_month_per_restaurant.flag=='Y']
```
