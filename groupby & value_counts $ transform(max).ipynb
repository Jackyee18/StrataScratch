# Find Favourite Wine Variety

Find each taster's favorite wine variety.
Consider that favorite variety means the variety that has been tasted by most of the time.
Output the taster's name along with the wine variety.


```python
import pandas as pd
import numpy as np
q=pd.read_csv('winemag_p2.csv',encoding='latin-1')
q
```

```python
q=q.groupby('taster_name').variety.value_counts().reset_index(name='total_tastings')
q
```

```python
q['max_total_tastings']=q.groupby('taster_name').total_tastings.transform(max)
q[q.total_tastings==q.max_total_tastings].iloc[:,0:3]
```
