# Common Friends Friend

Find the number of a user's friends' friend who are also the user's friend. Output the user id along with the count.




```python
import pandas as pd
import numpy as np
q=pd.read_csv('google_friends_network.csv')
q
```

```python
pd.merge(q,q,left_on='friend_id',right_on='user_id',how='left')
```
