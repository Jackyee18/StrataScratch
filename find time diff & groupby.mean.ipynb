# Session Type Duration


Calculate the average session duration for each session type?



```python
import pandas as pd
q=pd.read_table('twitch_sessions.txt')
q[['session_start','session_end']]=q[['session_start','session_end']].apply(pd.to_datetime)
q
```

```python
q['session_duration']=q.session_end-q.session_start
```

```python
q.groupby('session_type').session_duration.mean().reset_index(name='avg_duration')
```
