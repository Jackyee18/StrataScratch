# COMPLETED TRIP WITHIN 168 HOURS

An event is logged in the events table with a timestamp each time a new rider attempts a signup (with an event name 'attempted_su') or successfully signs up (with an event name of 'su_success').


For each city and date, determine the percentage of signups in the first 7 days of 2022 that completed a trip within 168 hours of the signup date. HINT: driver id column corresponds to rider id column

```python
import pandas as pd
import numpy as np
a=pd.read_csv('signup_events.csv')
b=pd.read_csv('trip_details.csv')
```

```python
# set option to display max rows of table
pd.set_option('display.max_rows', 500)
```

```python
# add the sign up time into the trip detail table
q=pd.merge(a,b,left_on='rider_id',right_on='driver_id').rename(columns={'timestamp':'signup_time'})
q
```

```python
# change type of date columns
q[['signup_time','request_at','actual_time_of_arrival']] = q[['signup_time','request_at','actual_time_of_arrival']].apply(pd.to_datetime)
# filter out with 'completed'
q=q[q['status']=='completed']
q
```

```python
# find the earliest time of trip for each rider
q['first_trip_time']=q.groupby('rider_id')['actual_time_of_arrival'].transform(min)
# find the timedelta between first trip and sign up time
q['first_vs_sign_up'] = q['first_trip_time'] - q['signup_time']
q
```

```python
# change timedelta into hour
q['completation_hour'] = q.first_vs_sign_up.dt.days * 24 + q.first_vs_sign_up.dt.seconds / 3600
```

```python
# find the ride who complete withiin 168 hours
q[q.completation_hour<=168].groupby('city_id_x')['rider_id'].count().reset_index()
q1=q[q.completation_hour<=168].groupby('city_id_x')['rider_id'].count().reset_index()
q1
```

```python
# create a table of the total riders per cities
qq=pd.merge(a,b,left_on='rider_id',right_on='driver_id').rename(columns={'timestamp':'signup_time'})
qq.groupby('city_id_x')['rider_id'].count().reset_index()
q2=qq.groupby('city_id_x')['rider_id'].count().reset_index()
q2
```

```python
# merge total riders of each cities with total complete riders of each cities
pd.merge(q1,q2,on='city_id_x').rename(columns={'rider_id_x':'total_complete_within_168','rider_id_y':'total_riders'})
q3=pd.merge(q1,q2,on='city_id_x').rename(columns={'rider_id_x':'total_complete_within_168','rider_id_y':'total_riders'})
q3
```

```python
# add a percentage column
q3['total_complete_within_168']/q3['total_riders']*100
q3['completing_percentage']=(q3['total_complete_within_168']/q3['total_riders']*100).round(0)
q3
```
