# Top 3 Wineries In The World


Find the top 3 wineries in each country based on the average points earned. In case there is a tie, order the wineries by winery name in ascending order. Output the country along with the best, second best, and third best wineries. If there is no second winery (NULL value) output 'No second winery' and if there is no third winery output 'No third winery'. For outputting wineries format them like this: "winery (avg_points)"

```python
import pandas as pd
import numpy as np
q=pd.read_csv('winemag_p1.csv',encoding='latin-1')
q
```

```python
q['ranking_point']=q.groupby('country').points.rank(method='dense',ascending=False)
```

```python
q1=q.loc[q.ranking_point==1,['id','country','points']]
```

```python
q2=q.loc[q.ranking_point==2,['id','country','points']]
```

```python
q3=q.loc[q.ranking_point==3,['id','country','points']]
```

```python
q=pd.merge(q1,q2,on='country',how='outer').merge(q3,on='country',how='outer').iloc[:,[1,0,2,3,4,5,6]]
q.rename(columns={'id_x':'best_wine_id','points_x':'best_wine_points','id_y':'second_wine_id',
                  'points_y':'second_wine_point','id':'third_wine_id','points':'third_wine_point'},inplace=True)
```

```python
q.loc[q.second_wine_id.isnull(),'second_wine_id']='No second winery'
q.loc[q.second_wine_point.isnull(),'second_wine_point']='No second winery points'
```

```python
q.loc[q.third_wine_id.isnull(),'third_wine_id']='No third winery'
q.loc[q.third_wine_point.isnull(),'third_wine_point']='No third winery points'
```
