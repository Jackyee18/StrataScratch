# Trips in Consecutive Months


Find the IDs of the drivers who completed at least one trip a month for at least two months in a row.



```python
import pandas as pd
import numpy as np
q=pd.read_csv('uber_trips.csv')
q.trip_date=pd.to_datetime(q.trip_date)
q=q[q.is_completed==True]
q=q.sort_values(['driver_id','trip_date'],ascending=[1,1])
q
```

```python
q['last_trip_date']=q.groupby('driver_id').trip_date.shift()
```

```python
q[q.trip_date.dt.month-q.last_trip_date.dt.month==1]
```

```python
q[q.trip_date.dt.month-q.last_trip_date.dt.month==-11]
```

```python
q[(q.trip_date.dt.month-q.last_trip_date.dt.month==1)|(q.trip_date.dt.month-q.last_trip_date.dt.month==-11)]
```

```python
q[(q.trip_date.dt.month-q.last_trip_date.dt.month==1)|(q.trip_date.dt.month-q.last_trip_date.dt.month==-11)].driver_id.unique()
```
