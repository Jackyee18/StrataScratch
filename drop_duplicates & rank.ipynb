# Distinct Salaries


Find the top three distinct salaries for each department. Output the department name and the top 3 distinct salaries by each department. Order your results alphabetically by department and then by highest salary to lowest.



```python
import pandas as pd
import numpy as np
q=pd.read_csv('twitter_employee.csv')
q
```

```python
# find the distinct values (salary of each department)
q=q.drop_duplicates(['department','salary'])
```

```python
# add the ranking salary per department
q['sal_ranking_per_dept']=q.groupby('department').salary.rank('min',ascending=False)
q['sal_ranking_per_dept']
```

```python
# find the top 3 distinct salary of each departments
q.loc[q.sal_ranking_per_dept<4,['department','salary']].sort_values(by=['department','salary'],ascending=[1,0])
```
