# Dates Of Inspection


Find the latest inspection date for the most sanitary restaurant(s). Assume the most sanitary restaurant is the one with the highest number of points received in any inspection (not just the last one). Only businesses with 'restaurant' in the name should be considered in your analysis.
Output the corresponding facility name, inspection score, latest inspection date, previous inspection date, and the difference between the latest and previous inspection dates. And order the records based on the latest inspection date in ascending order.

```python
import pandas as pd
import numpy as np
q=pd.read_csv('los_angeles_restaurant_health_inspections.csv')
pd.set_option('display.max_rows',500)
q['activity_date']=q['activity_date'].apply(pd.to_datetime)
q
```

```python
q=q.iloc[:,1:4].sort_values(by='activity_date',ascending=False)
```

```python
# find the previous score and inspection date groupby facility name: q.groupby('facility_name').shift(1)
# add the column of previous insepction date and previous score
q[['previous_inspection_date','previous_score']]=q.groupby('facility_name').shift(-1)
```

```python
# sort out the one which dont have the previous inspection date and previous insepction score
q=q[~q.previous_inspection_date.isnull()]
```

```python
# create the score difference between 2 inpection scores
q['score_diff']=q.score-q.previous_score
```
