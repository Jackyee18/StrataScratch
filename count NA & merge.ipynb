# Homework Results


Given the homework results of a group of students, calculate the average grade and the completion rate of each student. A homework is considered not completed if no grade has been assigned.
Output first name of a student, their average grade, and completion rate in percentages. Note that it's possible for several students to have the same first name but their results should still be shown separately.

```python
import pandas as pd
import numpy as np
q=pd.merge(pd.read_csv('allstate_homework.csv'),pd.read_csv('allstate_students.csv'))
q
```

```python
avg_grade=q.groupby('student_id').grade.mean().reset_index(name='avg_grade')
```

```python
q[q.grade.isnull()].groupby('student_id').student_id.count().reset_index(name='non_completion')
```

```python
q.groupby('student_id').student_id.count().reset_index(name='total_homeworks')
```

```python
completion=q.groupby('student_id').student_id.count().reset_index(name='total_homeworks').merge(q[q.grade.isnull()].groupby('student_id').student_id.count().reset_index(name='non_completion'),how='left').fillna(0)
completion
```

```python
completion['completion']=completion['total_homeworks']-completion['non_completion']
```

```python
completion['completion_per']=(completion.completion/completion.total_homeworks*100).round()
```

```python
completion
```

```python
final=pd.merge(avg_grade,completion)
final
```

```python
final[['student_id','avg_grade','completion_per']].merge(pd.read_csv('allstate_students.csv'))
```
